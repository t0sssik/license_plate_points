```python
%load_ext autoreload
%autoreload 2

import os
import sys
import matplotlib.pyplot as plt
from PIL import Image
import cv2

sys.path.append('..')
from utils.draw_points import draw_yolo_pose_points

supported_formats = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.tif')
```

## Кол-во изображений каждого класса

```python
root_dir = "../data/lpr_keypoints/RECT"
count_rect = 0
for root, dirs, files in os.walk(root_dir):
    for file in files:
        if file.endswith(supported_formats):
            count_rect += 1
            
root_dir = "../data/lpr_keypoints/SQUARE"
count_square = 0
for root, dirs, files in os.walk(root_dir):
    for file in files:
        if file.endswith(supported_formats):
            count_square += 1
            
print(f"Кол-во снимков однострочных номеров: {count_rect}")
print(f"Кол-во снимков двустрочных номеров: {count_square}")
```

## Min/max размеры изображений

```python
def get_image_size_pillow(image_path):
    with Image.open(image_path) as img:
        return img.size
    
root_dir = "../data/lpr_keypoints/"
widths = []
heights = []
for root, dirs, files in os.walk(root_dir):
    for file in files:
        if file.endswith(supported_formats):
            width, height = get_image_size_pillow(os.path.join(root, file))
            widths.append(width)
            heights.append(height)
            
print(f"Min ширина: {min(widths)}")
print(f"Max ширина: {max(widths)}")
print(f"Min высота: {min(heights)}")
print(f"Max высота: {max(heights)}")
```
