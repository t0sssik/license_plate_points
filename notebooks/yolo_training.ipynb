```python
from ultralytics import YOLO
```

```python
model = YOLO("yolo11n-pose.pt")
results = model.train(
    data="../data/data.yaml",
    epochs=500,
    imgsz=320,
    project="../runs/train",
    device=0,
    batch=-1,
    plots=True,
    save=True,
    verbose=True
)
```

```python
results = model.val(
    data="../data/data.yaml",
    split='val',  # используем валидационную выборку
    save=True,    # сохранять изображения с предсказаниями
    save_txt=True, # сохранять ключевые точки в txt
    save_json=True, # сохранять в JSON
    project="../runs/val",  # отдельная папка для валидации
    name="val_results"
)
```
